Ejercicio #1

```python
# función que convierte un número decimal a binario
def decimal_a_binario(numero):
    if numero == 0:
        return "0"
    binario = ""

    # el ciclo se repite mientras el número sea mayor que 0
    while numero > 0:
        residuo = numero % 2
        binario = str(residuo) + binario
        numero = numero // 2
    return binario


# pedimos al usuario que ingrese un número
num = int(input("Ingrese un número entero: "))

# mostramos el resultado de la conversión
print("El número en binario es:", decimal_a_binario(num))
```

Ejercicio #2

```python
# función que determina si un número es primo o no
def es_primo(numero):
    if numero <= 1:
        return False
    # probamos dividir el número entre todos los valores desde 2 hasta numero-1
    for i in range(2, numero):
        # si alguna división es exacta entonces no es primo
        if numero % i == 0:
            return False

    # si no encontramos divisores entonces el número es primo
    return True

num = int(input("Ingrese un número entero: "))

# imprimimos True o False dependiendo del resultado
print(es_primo(num))
```

Ejercicio #3

```python
import math

# función que calcula la desviación estándar de una lista
def desviacion_estandar(lista):

    # calcular el promedio de la lista
    promedio = sum(lista) / len(lista)

    # variable para acumular las diferencias al cuadrado
    suma = 0

    # recorrer cada número de la lista
    for num in lista:
        suma = suma + (num - promedio) ** 2

    # calcular la varianza
    varianza = suma / len(lista)

    # calcular la desviación estándar
    desviacion = math.sqrt(varianza)

    return desviacion


# lista de ejemplo
numeros = [10, 12, 23, 23, 16, 23, 21, 16]

# mostrar resultado
print(desviacion_estandar(numeros))
```

Ejercicio #4

```python
# función que ordena una lista usando bubble sort
def bubble_sort(lista):

    n = len(lista)

    # recorrer la lista varias veces
    for i in range(n):

        # comparar elementos vecinos
        for j in range(0, n - i - 1):

            # si el elemento actual es mayor que el siguiente se intercambian
            if lista[j] > lista[j + 1]:
                temp = lista[j]
                lista[j] = lista[j + 1]
                lista[j + 1] = temp

    return lista


# lista de ejemplo
numeros = [64, 34, 25, 12, 22, 11, 90]

# mostrar lista ordenada
print(bubble_sort(numeros))
```

Ejercicio #5

```python
# función que convierte una tupla en un string
def tupla_a_string(tupla):

    # variable donde se irá formando el string final
    resultado = ""

    # recorrer cada elemento de la tupla
    for elemento in tupla:
        resultado = resultado + str(elemento)

    # devolver el string completo
    return resultado


# tupla de ejemplo
datos = (10, 20, 40, 5, 70)

# mostrar el resultado
print(tupla_a_string(datos))
```

Ejercicio #6

```python
# función que elimina las tuplas vacías
def eliminar_vacias(lista):

    # lista donde guardaremos las tuplas que no estén vacías
    resultado = []

    # recorrer cada elemento de la lista
    for tupla in lista:

        # verificar que la tupla no esté vacía
        if tupla != ():
            resultado.append(tupla)

    return resultado


# lista de ejemplo
lista = [(), (), ('X',), ('a','b'), ('a','b','c'), ('d',)]

# mostrar resultado
print(eliminar_vacias(lista))
```

Ejercicio #7

```python
# función que calcula el promedio de cada tupla
def promedio_tuplas(tupla_datos):

    # lista temporal para guardar los promedios
    promedios = []

    # recorrer cada tupla interna
    for tupla in tupla_datos:

        # calcular el promedio de la tupla actual
        promedio = sum(tupla) / len(tupla)

        # guardar el promedio
        promedios.append(promedio)

    # convertir la lista final en tupla
    return tuple(promedios)


# tupla de entrada
datos = ((1, 1, 1, 12), (3, 45, 6, 5), (1, 8, 39, 32))

# mostrar resultado
print(promedio_tuplas(datos))
```

Ejercicio #8

```python
# definir la clase Libro
class Libro:

    # constructor para inicializar el libro con toda su información
    def __init__(self, titulo, autor, numero_paginas, leido):
        self.titulo = titulo
        self.autor = autor
        self.numero_paginas = numero_paginas
        self.leido = leido

    # método para marcar el libro como leído
    def marcar_leido(self):
        self.leido = True

    # método para marcar el libro como no leído
    def marcar_no_leido(self):
        self.leido = False

    # método para mostrar la información del libro
    def descripcion(self):

        if self.leido:
            estado = "Leído"
        else:
            estado = "No leído"

        return f"Título: {self.titulo}, Autor: {self.autor}, Páginas: {self.numero_paginas}, Estado: {estado}"


# crear libros
libro1 = Libro("The Great Gatsby", "F. Scott Fitzgerald", 516, True)
libro2 = Libro("1984", "George Orwell", 328, False)

# mostrar información
print(libro1.descripcion())
print(libro2.descripcion())
```

Ejercicio #9

```python
import math

# definir la clase Circulo
class Circulo:

    # constructor para inicializar el radio
    def __init__(self, radio):
        self.radio = radio

    # método para calcular el área
    def area(self):
        return math.pi * self.radio ** 2

    # método para calcular la circunferencia
    def circunferencia(self):
        return 2 * math.pi * self.radio


# pedir al usuario el radio
radio = float(input("Ingrese el radio del círculo: "))

# crear el objeto círculo
c1 = Circulo(radio)

# mostrar resultados
print("Área:", c1.area())
print("Circunferencia:", c1.circunferencia())
```
